```python
import matplotlib as mpl
#颜色和宽度
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.color'] = 'r'

import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
#设置线条颜色为红色
plt.rcParams['lines.color'] = 'r'
plt.plot(t,s)
c = np.cos(2*np.pi*t)
#设置线宽
plt.rcParams['lines.linewidth'] = '3'
plt.plot(t,c)
plt.show()
```

#### 箱线图解析
   * 箱体： 涵盖四分位数范围的矩形
   * 中值： 箱体中间的一条线
   * 箱须： 延伸到最大值和最小值的竖直的线，不包括异常值
   * 异常值：箱须之外的点



```python
import matplotlib.pyplot as plt
dataset = [113, 115, 119, 121, 124,
           124, 125, 126, 126, 126,
           127, 127, 128, 129, 130,
           130, 131, 132, 133, 136]
plt.subplot(121)
plt.boxplot(dataset, vert=False)
plt.subplot(122)
plt.hist(dataset) #直方图
plt.show()
```

### 正弦图 和 余弦图

![12](未命名截图.png)

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, 1000, endpoint=True)
y = np.cos(x)
y1 = np.sin(x)
plt.plot(x, y)
plt.plot(x, y1)
plt.title("Functions $\sin$ and $\cos$")
plt.xlim(-3.0, 3.0)
plt.ylim(-1.0, 1.0)
#$\$ 可在图中显示希腊字母
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           ['$-\pi$', '$-\pi/2$', '$0$', '$+\pi/2$', '$+\pi$'])
plt.yticks([-1, 0, +1],
         ['$-1$', '$0$', '$+1$'])
plt.show()
```
