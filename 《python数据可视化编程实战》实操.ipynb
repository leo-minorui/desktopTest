```python
import matplotlib as mpl
#颜色和宽度
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.color'] = 'r'

import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
#设置线条颜色为红色
plt.rcParams['lines.color'] = 'r'
plt.plot(t,s)
c = np.cos(2*np.pi*t)
#设置线宽
plt.rcParams['lines.linewidth'] = '3'
plt.plot(t,c)
plt.show()
```

### 正弦图 和 余弦图

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, 1000, endpoint=True)
y = np.cos(x)
y1 = np.sin(x)
plt.plot(x, y)
plt.plot(x, y1)
plt.title("Functions $\sin$ and $\cos$")
plt.xlim(-3.0, 3.0)
plt.ylim(-1.0, 1.0)
#$\$ 可在图中显示希腊字母
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           ['$-\pi$', '$-\pi/2$', '$0$', '$+\pi/2$', '$+\pi$'])
plt.yticks([-1, 0, +1],
         ['$-1$', '$0$', '$+1$'])
plt.show()
```

#### 设置坐标轴长度和范围

```python
import matplotlib.pyplot as plt

plt.axis()
#结果的值分别表示xmin xmax ymin ymax

l = [-1, 1, -10, 10]
plt.axis(l)
plt.axhline()
plt.axvline()
plt.axhline(4)
plt.grid(which='major')  #显示网格
plt.show()
```

#### 工作原理

> 如果不使用 `axis()` 或者其他参数设置, matplotlib会自动使用最小值，刚好可以让我们在一个图中看到所有的数据点。__如果设置`axis()`,其范围比数据集合中的最大值小，matplotlib会按照设置执行，这样就无法看到所有的数据点了__
>
>为避免这种情况发生 可以调用`autoscale()` (`matplotlib.pyplot.autoscale()`) 该方法会计算坐标轴的最佳大小以适应数据的显示
>`axhline() axcline()` 会根据给定x和y相应的绘制出相对于坐标轴的水平线和垂直线
>`axhline()`比较重要的参数是y向位置、xmin、xmax `axvline` 比较重要的参数是x向位置、ymin、ymax。

#### 设置刻度、刻度标签和网格

> 一个图形可以包括一个或多个子区
> 子区能以规则网格的方式排列plot

```python
import matplotlib.pyplot as plt
import numpy as np
# Get the current axes, creating one if necessary
ax = plt.gca()

# Set view to tight, and maximum number of tick intervals to 10
#nbins 需要分多少个间隔
ax.locator_params(tight=True, nbins = 10)

# generate 100 normal distribution values
ax.plot(np.random.normal(10, .1, 100))
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as dts
import datetime

fig = plt.figure()
# get current axis

ax = plt.gca()

#set some daterange
start = datetime.datetime(2013,1,1)
stop = datetime.datetime(2013,12,31)
delta = datetime.timedelta(days = 1) #表示时间间隔

#convert dates for matplotlib
dates = dts.drange(start, stop, delta)

#generate some random values

values = np.random.rand(len(dates))

#ax = plt.gca()
#create plot with dates
ax.plot_date(dates, values, linestyle='-', marker='')
#specify formater
date_format = dts.DateFormatter('%Y-%m-%d')
#apply formater
ax.xaxis.set_major_formatter(date_format)

fig.autofmt_xdate()
plt.show()

```
